**1.LOADING DATASET**

```python
import pandas as pd
```

```python
df=pd.read_csv('CarPrice_Assignment.csv')
print("Dataset Preview")
df
```

```python
df.info()
```

```python
df.describe()
```

**2.INPUT FEATURE AND OUTPUT VARAIBLE**

```python
#independent varaible
x=df[['enginesize']]
```

```python
#dependent variable
y=df['price']
```

```python
#training he model using Linear Regression
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

```python
#training the model using Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

**3.VISUALIZATION**

```python
import matplotlib.pyplot as plt
```

```python
#Visualising the Linear Regression results
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
plt.title('Linear Regression(Engine Size vs Price)')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.show()
```

```python
#Visualising the Polynomial Regression results
import numpy as np
x_range = np.linspace(x.min().iloc[0], x.max().iloc[0], 500).reshape(-1, 1)
x_range_df = pd.DataFrame(x_range, columns=['enginesize'])
y_range_pred = lin_reg_2.predict(poly_reg.transform(x_range_df))
plt.scatter(x, y, color='red')
plt.plot(x_range_df, y_range_pred, color='blue')
plt.title('Polynomial Regression (Engine Size vs Price)')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.show()
```

**4.PREDICTIVE MODEL TO ESTIMATE CAR PRICE BASED ON ENGINE SIZE**

```python
#linear reg
y_pred_lin = lin_reg.predict(x)
```

```python
#polynomial reg
y_pred_poly = lin_reg_2.predict(X_poly)
```

**5.PREDICT PRICE OF CAR WITH ENGINE SIZE=200**

```python
import pandas as pd

engine_size_df = pd.DataFrame({'enginesize':[200]})

# Linear Regression
price_lin = lin_reg.predict(engine_size_df)
print(f"Predicted price using Linear Regression: {price_lin[0]:.2f}")

# Polynomial Regression
price_poly = lin_reg_2.predict(poly_reg.transform(engine_size_df))
print(f"Predicted price using Polynomial Regression: {price_poly[0]:.2f}")
```

**6.PERFORMANCE ANALYSIS**

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
#for linear regression
mae=mean_absolute_error(y,y_pred_lin)
mse=mean_squared_error(y,y_pred_lin)
r2=r2_score(y,y_pred_lin)
rmse=np.sqrt(mse)
print("Linear Regression Performance Metrics:")
print("Mean Absolute Error: ",mae)
print("Mean Squared Error: ",mse)
print("Root Mean Squared Error: ",rmse)
print("R-squared: ",r2)
```

```python
#for polynomial regression
mae=mean_absolute_error(y,y_pred_poly)
mse=mean_squared_error(y,y_pred_poly)
r2=r2_score(y,y_pred_poly)
rmse=np.sqrt(mse)
print("Polynomial Regression Performance Metrics:")
print("Mean Absolute Error: ",mae)
print("Mean Squared Error: ",mse)
print("Root Mean Squared Error: ",rmse)
print("R-squared: ",r2)
```
